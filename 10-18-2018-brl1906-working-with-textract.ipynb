### <font color=orange>strenghts of textract</font>
**Using options pdf & tesseract method return good results when the PDF is just text and not a table.  Seems optimized for raw text documents, returning entire document**

```python
# using the pdf extension and tesseract options handles text only (no tables) very well
textract.process('data/pdf_data/epa-letter.pdf', extension='pdf', method='tesseract').decode()
```
